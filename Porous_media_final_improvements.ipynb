```python
import numpy as np
import matplotlib.pyplot as plt
figure,axes=plt.subplots()
centerX=[]
centerY=[]
radius=[]
X=np.random.uniform(0,1)
Y=np.random.uniform(0,1)
r=np.random.uniform(.02,.1)
centerX.append(X)
centerY.append(Y)
radius.append(r)
Number_of_circles=0
for i in range(2000):
    X=np.random.uniform(0,1)
    Y=np.random.uniform(0,1)
    r=np.random.uniform(.02,.2)
    count=0

    for j in range(len(centerX)-1):
        
        if np.sqrt((X-centerX[j])**2+(Y-centerY[j])**2) <(radius[j]+r):
            
            count+=1
            
    if count==0:
        
        Circle = plt.Circle((X, Y), r, fill=False)
        axes.set_aspect(1)
        axes.add_artist(Circle)
        centerX.append(X)
        centerY.append(Y)
        radius.append(r)
        Number_of_circles+=1

print('The total number of circle placed={}'.format(Number_of_circles))       
plt.show()


```
